```python
for title, vote , votecount in zip(data['original_title'], data['vote_average'], data['vote_count']):
    if vote >= 8:
        print(title,' -: ',(vote*votecount/100))
```

```python
movies = [
    (title, vote, vote * votecount / 100)
    for title, vote, votecount in zip(data['original_title'], data['vote_average'], data['vote_count'])
    if vote >= 8
]

# Sort movies by the calculated score in descending order
movies.sort(key=lambda x: x[2], reverse=True)

# Print the sorted list with a clear pattern
for title, vote, score in movies:
    print(f"{title}: Vote Average = {vote}, Score = {score:.2f}")
```
